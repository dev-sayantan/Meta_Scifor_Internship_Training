## Pandas DataFrame

<a id="5"></a> <br>
#### Dictionary Creation

```python
# dictionary creation
dict = {"Name":['Sayantan','Sudip','Soumya','Pradeep'],"Qualification":['BSc','MTech','MBBS','MSc'],"GPA":[6.57,7.9,8.6,7],"Univ":['MAKAUT','JIS','MMC&H','KIIT']}
student_details = pd.DataFrame(dict)
student_details
```

<a id="6"></a>
#### Adding New Data Column With Values

```python
# now adding a new data field(column) and also it's corresponding records(rows).
student_details['Location'] = "Kolkata,Kalyani,Midnapore,Bhubaneshwar".split(',')
student_details
```

<a id="7"></a>
#### Adding New Column With INSERT Method

```python
## applying insert() method we can add column to the dataframe, alongwith values, on a specific index position

student_details.insert(2,'Stream',['Biotech','EIE','Orthopedics','Genetics'])
student_details
```

<a id="8"></a>
#### Creating Columns with NaN Values

```python
# WORKING WITH NaN(Not a Number)Values
student_details.insert(6,'Submitted Projects',[2,np.nan,np.nan,3])
student_details
```

<a id="9"></a>
#### Drop Columns With NaN Values

```python
student_details.insert(1,'Last Name',[np.nan,np.nan,np.nan,np.nan])
student_details
```

```python
print(student_details.dropna(axis=1)) ## eliminating columns with NaN values
## Here we didn't see the change reflected but when we print the result, it reflects
print('--'*50)
print(student_details)
```

<a id="10"></a>
#### Filling NaN Values

```python
# Filling the NaN values with a defined value
print(student_details)
student_details.fillna(value='-')
```

<a id="14"></a>
#### Concatinating DataFrame

```python
# CONCATINATING DATAFRAME --> There are three main methods, 1.Contact 2.Merge 3.Join
student_details_1 = student_details.copy()
student_details_1
```

```python
dict_1 = {"Name":['Sourav','Suparna','Amit','Debjit'],"Last Name":[np.nan,np.nan,np.nan,np.nan],"Qualification":['BSc','BTech','MCA','MSc'],"Stream":['Zoology','Bioinformatics','Cloud Computing','Math'],"GPA":[5.7,8.9,6.7,10],"Univ":['MAKAUT','PES','Jain Univ','NBU'],"Location":['Kolkata','Bengaluru','Bengaluru','Siliguri'],"Submitted Projects":[3,6,2,4]}
student_details_2 = pd.DataFrame(dict_1)
student_details_2
```

```python
dict_2 = {"Name":['Saroj','Amrita','Mukesh','Arijit'],"Last Name":[np.nan,np.nan,'Kumar','Ghosh'],"Qualification":['MA','BTech','BCA','BSc'],"Stream":['Psychology','Bioinformatics','FullStack Dev','Physics'],"GPA":[7.2,9.1,6,8],"Univ":['WBSU','TIU','Shoolini','JU'],"Location":['Barasat','Kolkata','Solan','Jadavpur'],"Submitted Projects":[3,3,4,5]}
student_details_3 = pd.DataFrame(dict_2)
student_details_3
```

<a id="15"></a>
##### CONCAT Row-wise

```python
# CONCAT ROW-WISE
students_details_all = pd.concat([student_details_1,student_details_2,student_details_3],axis=0)
students_details_all
```

<a id="16"></a>
##### CONCAT Column-wise

```python
# CONCAT COLUMN-WISE
students_details_all_concatCol = pd.concat([student_details_1,student_details_2,student_details_3],axis=1)
students_details_all_concatCol
```

```python
# Filling all NaN Values with Zero(0)
students_details_all.fillna(value=0,inplace=True)
students_details_all
```

<a id="17"></a>
##### MERGE

```python
# MERGE
players = {'Name':['Sachin','Sourav','Dravid','Laxman','Yuvraj','Zaheer'],'Runs':[10234,6786,8990,7890,6700,1200],'Wickets':[100,78,2,0,123,345]}
player_details = pd.DataFrame(players)
player_details
```

```python
players2 = {'Name':['Sachin','Sourav','Dravid','Laxman','Yuvraj','Zaheer'],'Catches':[120,90,200,130,190,70]}
player_details2 = pd.DataFrame(players2)
player_details2
```

```python
print("====== DataSet is Merged based on the 'Name' field ==========")
pd.merge(player_details,player_details2,on='Name')
```

##### JOIN : Inner JOIN

```python
left_df = pd.DataFrame({'A':['A0','A1','A2','A4'],'B':['B0','B1','B2','B4'],'C':['C0','C1','C2','C4'],'D':['D0','D1','D2','D4']},index=['I0','I1','I2','I3'])
right_df = pd.DataFrame({'E':['E0','E1','E2','E4'],'F':['F0','F1','F2','F4'],'G':['G0','G1','G2','G4'],'H':['H0','H1','H2','H4']},index=['I0','I1','I2','I4'])
```

```python
left_df
```

```python
right_df
```

```python
# Joinning both the dataframes
inner_join = left_df.join(right_df)
```

<a id="19"></a>
##### JOIN : Outer JOIN

```python
# Joinning both the dataframes by outer-join
left_df.join(right_df,how='outer')
```
